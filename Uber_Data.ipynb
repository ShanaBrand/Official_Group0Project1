```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib
#from mpl_toolkits.basemap import Basemap
```

```python
DATA_FILE = 'Uber_Clean.csv'
uber_data = pd.read_csv(DATA_FILE)
uber_data.head()
```

```python
uber_data['Date'] = pd.to_datetime(uber_data['Date'], format="%Y-%m-%d")
uber_data['Time'] = pd.to_datetime(uber_data['Time'], format="%H:%M:%S")
uber_data['DayOfWeek'] = uber_data['Date'].dt.dayofweek
uber_data['HourOfDay'] = uber_data['Time'].dt.hour
```

```python
uber_monthdays = uber_data.pivot_table(index=['DayOfWeekNum'],
                                  values='borough',
                                  aggfunc='count')
uber_monthdays.plot(kind='bar', figsize=(8,6))
plt.ylabel('Total Journeys')
plt.title('Journeys by Month Day');
```

```python
uber_monthdays = uber_data.pivot_table(index=['Day'],
                                  values='borough',
                                  aggfunc='count')
uber_monthdays.plot(kind='bar', figsize=(8,6))
plt.ylabel('Total Journeys')
plt.title('Journeys by Month Day');
```

```python
uber_hour = uber_data.pivot_table(index=['HourOfDay'],
                                  values='borough',
                                  aggfunc='count')
uber_hour.plot(kind='bar', figsize=(8,6))
plt.ylabel('Total Journeys')
plt.title('Journeys by Hour');
```

```python
uber_monthdays = uber_data.pivot_table(index=['borough'],
                                  values='service_zone',
                                  aggfunc='count')
uber_monthdays.plot(kind='bar', figsize=(8,6))
plt.ylabel('Total Journeys')
plt.title('Journeys by Month Day');
```
