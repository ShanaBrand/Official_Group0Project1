```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
DATA_FILE = 'NYCTaxiTrips_Yr2015.csv'
uber_data_0 = pd.read_csv(DATA_FILE)
uber_data_0.head()
```

```python
uber_data = uber_data_0
uber_data['pickup_datetime'] = pd.to_datetime(uber_data['pickup_datetime'], format="%Y-%m-%d %H:%M:%S")
uber_data['DayOfWeekNum'] = uber_data['pickup_datetime'].dt.dayofweek
uber_data['DayOfWeek'] = uber_data['pickup_datetime'].dt.weekday_name
uber_data['MonthDayNum'] = uber_data['pickup_datetime'].dt.day
uber_data['HourOfDay'] = uber_data['pickup_datetime'].dt.month
```

```python
#Scatter(Trip_Distance vs Tip_Amount)
x = uber_data["trip_distance"]
y = uber_data["tip_amount"]
plt.ylabel('Tip Amount($)')
plt.xlabel('Trip Distance(miles)')
plt.title('Tip Amount by Trip Distance')
plt.xlim(0,100)
plt.ylim(0,150)
plt.xlim(0,40)
plt.ylim(0,40)
plt.scatter(x,y,s=1)
plt.show()
```

```python
fig, axes = plt.subplots()
#figure two
x2 = uber_data[["pickup_zip","tip_amount"]]
x3 = uber_data["pickup_zip"].value_counts().reset_index()
groupby = x2.groupby("pickup_zip")
x4 = groupby.mean()

x4["tip_count"] = groupby.count()
#sd = np.std(x4["tip_amount"], axis=0)
x5 = x4.loc[x4["tip_count"]>20,]
x6 = x5.loc[x5["tip_amount"]>1.6,].reset_index()
x = x6["pickup_zip"]
y = x6["tip_amount"]
y_pos = np.arange(len(x))

axes.barh(y_pos, y, align='center', alpha=0.5)
plt.yticks(y_pos, x)
axes.set_xlabel('Tip Amount')
plt.title('Pickup Zip Code')
```
